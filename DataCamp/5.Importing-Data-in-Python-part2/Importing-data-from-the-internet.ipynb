# Importing flat files from the web
* Imported file from the web and save it locally and loaded into DF
    * url of the import file: 'https://s3.amazonaws.com/assets.datacamp.com/production/course_1606/datasets/winequality-red.csv'


```python
from urllib.request import urlretrieve
import pandas as pd
url = 'https://s3.amazonaws.com/assets.datacamp.com/production/course_1606/datasets/winequality-red.csv'
urlretrieve(filename='winequality-red.csv', url=url)
df = pd.read_csv('winequality-red.csv', sep=';')
print(df.head())
```

# Opening and reading flat files from the web
* Load file direct into DF from the web using **pandas**
    * use **pd.read_csv()**
    * url same as before

```python
import matplotlib.pyplot as plt
import pandas as pd
url = 'https://s3.amazonaws.com/assets.datacamp.com/production/course_1606/datasets/winequality-red.csv'
df = pd.read_csv(url, sep=';')
print(df.head())
pd.DataFrame.hist(df.iloc[:, 0:1])
plt.xlabel('fixed acidity (g(tartaric acid)/dm$^3$)')
plt.ylabel('count')
plt.show()
```

# Performing HTTP requests in Python using urllib

* Ping Datacamp serves to perform GET request to extract info. from tech page: "http://www.datacamp.com/teach/documentation"

```python
from urllib.request import urlopen, Request
url = "http://www.datacamp.com/teach/documentation"
request = Request(url)
response = urlopen(request)
print(type(response))
response.close()
```

# Printing HTTP request results in Python using urllib

* Deal with the object: http.client.HTTPResponse
    * Came from HTML
    * read to extract the HTML using **read()**
    

```python
from urllib.request import urlopen, Request
url = "http://www.datacamp.com/teach/documentation"
request = Request(url)
response = urlopen(request)
html = response.read()
print(html)
response.close()
```

# Performing HTTP requests in Python using requests

```python
import requests
url = "http://www.datacamp.com/teach/documentation"
r = requests.get(url)
text = r.text
print(text)
```

# Parsing HTML w/ beautifulSoup

* BeautifulSoup pkg
    * parse, prettify and extra info. from HTML
    * url: url = 'https://www.python.org/~guido/'

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.python.org/~guido/'
r = requests.get(url)
html_doc = r.text
soup = BeautifulSoup(html_doc)
pretty_soup = soup.prettify()
print(pretty_soup)
```

# Turning a webpage into Data using BeautifulSoup: getting the text
* Extract text from the BDFL's webpage

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.python.org/~guido/'
r = requests.get(url)
html_doc = r.text
soup = BeautifulSoup(html_doc)
guido_title = soup.title
print(guido_title)
guido_text = soup.text
print(guido_text)
```

# Getting the hyperlinks

* **find_all()**

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.python.org/~guido/'
r = requests.get(url)
html_doc = r.text
soup = BeautifulSoup(html_doc)
print(soup.title)
a_tags = soup.find_all(name='a')
for link in a_tags:
    print(link.get('href'))
```
